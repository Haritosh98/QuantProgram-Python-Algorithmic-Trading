## 2.4 Focusing on AAPL

```python
aapl = close["AAPL"].copy().to_frame()
aapl
```

```python
# calculate the increase or decease from every day to the next
# using the shift function (which shifts the values down one step)

aapl.shift(periods = 1)
```

```python
aapl['lag1'] = aapl.shift(periods = 1)
```

```python
aapl
```

```python
# now want to find the daily change so w can subtract the lag1 from the original column

aapl['diff'] = aapl['AAPL'] - aapl['lag1']
aapl
```

```python
# now we can calculate the daily percentage change
aapl['%diff'] = aapl['diff'].div(aapl['lag1']).mul(100)
aapl
```

```python
# another way to do this is to use the diff method
aapl['diff_2'] = aapl['AAPL'].diff(periods = 1)

# thus we can calculate the percentage change
aapl['%diff_2'] = aapl['diff_2'].div(aapl['AAPL'].shift(periods = 1)).mul(100)
aapl
```

```python
# we can also use the pct_change method
aapl['%diff_3'] = aapl['AAPL'].pct_change().mul(100)
aapl
```

```python
# now delete the columns we don't need
aapl = aapl.drop(columns = ['lag1', 'diff', '%diff', 'diff_2', '%diff_2'])
```

```python
# and then rename the columns
aapl.columns = ['AAPL', 'daily_pct_change']
aapl
```

```python
# now converting the daily data to monthly data (for example the last day of the month)

aapl['AAPL'].resample('M').last()
```

```python
# for the last business day of the month
aapl['AAPL'].resample('BM').last()
```

```python
# not calculating the monthly percentage change for the last business day of the month
monthly = aapl['AAPL'].resample('BM').last().pct_change().mul(100).to_frame()
monthly
```

### 2.4.1 Now calculating the mean and the standard deviation

```python
# now we are going to calculate the return as daily percentage changes:

# redefining apple daily closing prices
aapl = stocks.loc[:,"Close"]["AAPL"].to_frame()
aapl
```

```python
daily_return = aapl.pct_change().mul(100).dropna()
daily_return
```

```python
daily_return.info()
```

### 2.4.2 Producing the plot for this

```python
daily_return.plot(kind = "hist", figsize=(12,8), bins=100, fontsize = 15)
plt.show()
# the graph suggests that the daily returns are normally distributed
# we can also see that the mean is positive, which means that on average the stock goes up
# However, we can't base exit strategies on this, because the stock can go up 5% and then go down 10% the next day.
```

### 2.4.3 calculating the mean, variance and standard deviation

```python
daily_mean_ret = daily_return.mean()
daily_mean_ret
```

```python
daily_var_ret = daily_return.var()
daily_var_ret
```

```python
# the standard deviation is the square root of the variance
# this is call the risk metric of the stock

daily_std_ret = daily_return.std()
daily_std_ret
```

### 2.4.4 The annual mean, variance and standard deviation for the return

```python
annual_mean_ret = daily_mean_ret.mul(252) # since there are 252 trading days in a year
annual_mean_ret
```

```python
annual_variance_ret = daily_var_ret.mul(252)
annual_variance_ret
```

```python
annual_std_ret = np.sqrt(annual_variance_ret)
annual_std_ret
```

## 2.5 Now doing this for the rest of the data

```python
close = stocks.loc[:,"Close"].copy()
close
```

```python
norm_close = close.div(close.iloc[0]).mul(100)
norm_close
```

```python
daily_ret = close.pct_change().mul(100).dropna()
daily_ret
# this calculates the % change for all the stocks in the list (tickers) on a daily basis and drops the first row (which is NaN) and multiplies by 100 to get the % change 
```

```python
daily_ret.describe()
# this gives us the mean, std, min, max, etc. for all the stocks in the list (tickers) on a daily basis  
```

```python
# now transposing this data to get the mean, std, etc. for each stock

daily_ret.describe().T
```

```python
summary = daily_ret.describe().T.loc[:, ['mean', 'std']]
summary
```

```python
# the above is for the daily returns, now we want to calculate the annual returns
annual_summary = summary.copy()
annual_summary['mean'] = summary['mean'].mul(252)
annual_summary['std'] = summary['std'].mul(np.sqrt(252))  
annual_summary
```

### 2.5.1 Plotting this

```python
annual_summary.plot(kind = 'scatter', x = 'std', y = 'mean', figsize=(12,8), fontsize=12)
```

```python
# now annotate the scatter plot with the stock names
annual_summary.plot(kind = 'scatter', x = 'std', y = 'mean', figsize=(12,8), fontsize=12)
for i in annual_summary.index:
    plt.annotate(i, xy=(annual_summary.loc[i,'std'] + 0.002, annual_summary.loc[i,'mean'] + 0.002), size=15)
plt.xlabel('Annual Risk (Standard Deviation %)', fontsize=15)
plt.ylabel('Annual Return (Mean %)', fontsize=15)
plt.title('Risk vs. Return', fontsize=15)
```

```python
# returns on apple is high but so is the risk
# also choosing between microsoft and amazon is a tradeoff between risk and return
# choosing between disney and microsoft - choose microsoft because it has higher returns at almost the same risk
# choosing between goog and intc - choose goog because it has higher returns at almost the same risk
# it looks like tech stocks are the best to invest in because they have the highest returns at the same risk

# this is basically what warren buffet does - he chooses stocks that have high returns and low risk
```
